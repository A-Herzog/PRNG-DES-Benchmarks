# Comparision of different PRNGs (Model 1)

The Python code in this notebook will load the simulation results stored in the `statistics` folder and will generate the tables for model 1 shown in the report.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), stats.sem(a)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h, h
```

## Load and process data

```python
data = [
    {"name": "Exponential", "file": "statistics/results1Exp.txt", "mean": 200, "std": 200},
    {"name": "Logarithmic", "file": "statistics/results1Log.txt", "mean": 200, "std": 50},
    {"name": "Gamma", "file": "statistics/results1Gamma.txt", "mean": 200, "std": 50},
    {"name": "Triangular", "file": "statistics/results1Triangular.txt", "mean": 200, "std": 50}
]
```

```python
for i in range(len(data)):
    data[i]["df"] = pd.read_csv(data[i]["file"], header=None, names=["rng_mode", "n", "mean", "std", "delta2", "time"], sep="\t")
```

```python
generators=data[0]["df"]["rng_mode"].drop_duplicates().tolist()
generators.remove("ThreadLocalRandomSlow")  # This is the same a "ThreadLocalRandom". Only used for speed comparision. Removed from other tables.
```

```python
for rec in data:
    df = rec["df"]
    mean = rec["mean"]
    std = rec["std"]

    mean_error = []
    std_error = []
    delta2 = []
    time = []

    for generator in generators:
        mean_error.append(round(abs(df[df["rng_mode"] == generator]["mean"].mean() - mean) / mean * 100, 5))  # Values are in percentage
        std_error.append(round(abs(df[df["rng_mode"] == generator]["std"].mean() - std) / std * 100, 5))  # Values are in percentage
        delta2.append(round(df[df["rng_mode"] == generator]["delta2"].mean() * 100000000, 5))
        time.append(round(df[df["rng_mode"] == generator]["time"].mean(), 5))

    rec["mean_error"] = mean_error
    rec["std_error"] = std_error
    rec["delta2"] = delta2
    rec["time"] = time
```

### Confidence interval for mean

```python
conf = {}
h_list = []
for rec in data:
    name = rec["name"]
    df = rec["df"]
    for generator in generators:
        l = list(df[df["rng_mode"] == generator]["mean"])
        h = mean_confidence_interval(l)[3]  # 95%
        if name not in conf: conf[name] = []
        conf[name].append(h)
        h_list.append(h)

df_conf = pd.DataFrame(conf, index=generators)
df_conf
```

```python
print(df_conf[~df_conf.index.isin(["Drand48", "Drand48Mix"])].max())

print(df_conf[df_conf.index.isin(["Drand48", "Drand48Mix"])].max())
```

## Times

```python
generators.append("ThreadLocalRandomSlow")  # Not in quality comparison list since only used for speed comparision

for rec in data:
    df = rec["df"]
    mean = rec["mean"]
    std = rec["std"]

    mean_error = []
    std_error = []
    delta2 = []
    time = []

    for generator in generators:
        mean_error.append(round(abs(df[df["rng_mode"] == generator]["mean"].mean() - mean) / mean * 100, 5))  # Values are in percentage
        std_error.append(round(abs(df[df["rng_mode"] == generator]["std"].mean() - std) / std * 100, 5))  # Values are in percentage
        delta2.append(round(df[df["rng_mode"] == generator]["delta2"].mean() * 100000000, 5))
        time.append(round(df[df["rng_mode"] == generator]["time"].mean(), 5))

    rec["mean_error"] = mean_error
    rec["std_error"] = std_error
    rec["delta2"] = delta2
    rec["time"] = time
```

```python
average_time = [sum([data[i]["time"][j] for i in range(len(data))]) for j in range(len(generators))]

d = {}
for rec in data:
    d[rec["name"]] = rec["time"]/min(rec["time"])*100

d["Average"] = average_time/np.min(average_time)*100

df_time = pd.DataFrame(d, index=generators)

df_time
```
