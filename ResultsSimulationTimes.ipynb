```python
data1 = [
    # Model 1a
    {"file": "statistics/results1Exp.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Log.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Gamma.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Triangular.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Poisson.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Geometric.txt", "countCol": 1, "timeCol": -1},
    # Model 1b
    {"file": "statistics/results1ExpAutocorrelation.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1LogAutocorrelation.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1GammaAutocorrelation.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1TriangularAutocorrelation.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1PoissonAutocorrelation.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1GeometricAutocorrelation.txt", "countCol": 1, "timeCol": -1},
    # Model 2a
    {"file": "statistics/results2MM1rho90.txt", "timeCol": -1},
    {"file": "statistics/results2MM1rho95.txt", "timeCol": -1},
    # Model 2b
    {"file": "statistics/results2MG1LogRho90.txt", "timeCol": -1},
    {"file": "statistics/results2MG1LogRho95.txt", "timeCol": -1},
    {"file": "statistics/results2MG1GammaRho90.txt", "timeCol": -1},
    {"file": "statistics/results2MG1GammaRho95.txt", "timeCol": -1},
    # Model 3a
    {"file": "statistics/results3a.txt", "timeCol": -1},
    # Model 3b
    {"file": "statistics/results3b.txt", "timeCol": -1},
]

data2 = [
    # Model 2b - rho
    {"file": "statistics/results2MG1Log-variatingRho.txt", "timeCol": -1},
]

data3arm = [
    # Model 1a - Raspberry Pi
    {"file": "statistics/results1Exp-raspberrypi.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Log-raspberrypi.txt", "countCol": 1, "timeCol": -1},
    # {"file": "statistics/results1Gamma-raspberrypi.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Triangular-raspberrypi.txt", "countCol": 1, "timeCol": -1},
    # {"file": "statistics/results1Poisson-raspberrypi.txt", "countCol": 1, "timeCol": -1},
    # {"file": "statistics/results1Geometric-raspberrypi.txt", "countCol": 1, "timeCol": -1},
]

data3intel = [
    # Model 1a - Intel Core Ultra 5 225H
    {"file": "statistics/results1Exp-intel.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Log-intel.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Gamma-intel.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Triangular-intel.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Poisson-intel.txt", "countCol": 1, "timeCol": -1},
    {"file": "statistics/results1Geometric-intel.txt", "countCol": 1, "timeCol": -1},
]
```

```python
def calculate_total_hours(data):
    time = 0
    for rec in data:
        df = pd.read_csv(rec["file"], header=None, sep="\t")
        if rec["timeCol"] > 0:
            time += sum(df[rec["timeCol"]])
        else:
            cols = len(df.columns)
            col = cols + rec["timeCol"]
            time += sum(df[col])
    return time / 1000 / 3600  # Convert from milliseconds to hours

def calculate_total_count(data):
    count = 0
    for rec in data:
        if not "countCol" in rec:
            continue
        df = pd.read_csv(rec["file"], header=None, sep="\t")
        count += sum(df[rec["countCol"]])
    return count

def get_lines_count(data):
    lines = 0
    for rec in data:
        if "countCol" in rec:
            continue
        df = pd.read_csv(rec["file"], header=None, sep="\t")
        lines += len(df)
    return lines
```

## Simulated arrivals

```python
print("Model 1a, 1b:")
print(round(calculate_total_count(data1)/1e12, 2), "* 10^12 simulations")
print("Model 2a, 2b, 3a, 3b:")
print(get_lines_count(data1)/1e4, "* 10^12 simulations")
print("Model 2b - rho:")
print(get_lines_count(data2)/1e3, "* 10^12 simulations")
```

## Simulation time in hours

```python
print("Model 1a, 1b, 2a, 2b, 3a, 3b:")
print(round(calculate_total_hours(data1),1),"hours")
print("Model 2b - rho:")
print(round(calculate_total_hours(data2),1),"hours")
print("Model 1a - Raspberry Pi:")
print(round(calculate_total_hours(data3arm),1),"hours")
print("Model 1a - Intel Core Ultra 5 225H:")
print(round(calculate_total_hours(data3intel),1),"hours")
```
