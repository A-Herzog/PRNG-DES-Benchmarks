# Summation of the needed simulation time

```python
import pandas as pd
```

```python
data = [
    # Model 1a
    {"file": "statistics/results1Exp.txt", "col": -1},
    {"file": "statistics/results1Log.txt", "col": -1},
    {"file": "statistics/results1Gamma.txt", "col": -1},
    {"file": "statistics/results1Triangular.txt", "col": -1},
    {"file": "statistics/results1Poisson.txt", "col": -1},
    {"file": "statistics/results1Geometric.txt", "col": -1},
    # Model 1b
    {"file": "statistics/results1ExpAutocorrelation.txt", "col": -1},
    {"file": "statistics/results1LogAutocorrelation.txt", "col": -1},
    {"file": "statistics/results1GammaAutocorrelation.txt", "col": -1},
    {"file": "statistics/results1TriangularAutocorrelation.txt", "col": -1},
    # Model 2a
    {"file": "statistics/results2MM1rho90.txt", "col": -1},
    {"file": "statistics/results2MM1rho95.txt", "col": -1},
    # Model 2b
    {"file": "statistics/results2MG1rho90.txt", "col": -1},
    {"file": "statistics/results2MG1rho95.txt", "col": -1},
    # Model 3a
    {"file": "statistics/results3a.txt", "col": -1},
    # Model 3b
    {"file": "statistics/results3a.txt", "col": -1},
]
```

```python
time = 0
for rec in data:
    df = pd.read_csv(rec["file"], header=None, sep="\t")
    if rec["col"] > 0:
        time += sum(df[rec["col"]])
    else:
        cols = len(df.columns)
        col = cols + rec["col"]
        time += sum(df[col])
```

## Simulation time in hours

```python
time / 1000 / 3600
```
